# Customer Sales Analysis
This notebook performs a full analysis of the customer sales dataset, including:
- Data loading
- Cleaning
- Monthly sales analysis
- Top customers
- Category performance
- Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data.csv')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
# Monthly Sales
monthly = df.groupby(df['date'].dt.to_period('M'))['amount'].sum()
monthly
```

```python
# Plot Monthly Sales
plt.figure()
monthly.plot(kind='bar')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Top Customers
top = df.groupby('customer_name')['amount'].sum().sort_values(ascending=False)
top
```

```python
# Category Sales
cat = df.groupby('category')['amount'].sum()
cat
```
